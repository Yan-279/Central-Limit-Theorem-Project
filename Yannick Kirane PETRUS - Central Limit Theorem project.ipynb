```python
%matplotlib inline
import numpy as np
import random
import matplotlib.pyplot as plt

#1) Creating L1 simulating the toss of a coin n times

n = int(input("Enter number of tosses\n"))
L1 = np.array([np.random.randint(0,2,n)])
#print(L1)


#2) Creating the list L2 with N values taking a sample of random size results among L1 and making the sum

N = int(input("\nEnter the number of samples N\n"))
size = int(input("\nEnter the size of the samples size\n"))
randomlist = [random.sample(range(0, n), size) for i in range (N)]
#print("\nrandomlist\n" , randomlist)
L2 = [sum(L1[0,randomlist[i]]) for i in range (len(randomlist))]
avgL2 = [np.average(L1[0,randomlist[i]]) for i in range (len(randomlist))]
#print(L2)
#print(avgL2)


#3) Representing this last list L2 as a histogram

L2 = np.array(L2)

sigma = np.std(L2)
mu = n
print("\nmu: ",mu)         # mean of distribution
print("sigma: ",sigma)     # standard deviation of distribution

sample = mu + sigma * L2

plt.hist(sample);

```

```python
#Putting the previous instructions inside a function:
def toss(n,N,size):
    L1 = np.array([np.random.randint(0,2,n)])
    randomlist = [random.sample(range(0, n), size) for i in range (N)]
    L2 = [sum(L1[0,randomlist[i]]) for i in range (len(randomlist))]
    avgL2 = [np.average(L1[0,randomlist[i]]) for i in range (len(randomlist))]
    L2 = np.array(L2)
    sigma = np.std(L2)
    mu = n
    sample = mu + sigma * L2

    plt.hist(sample);
    
toss(10000,1000,1000)
```

```python
#Let's try changing "N", the number of times we take samples to see the difference
toss(10000,100,1000)
```

```python
toss(10000,50,1000)
```

```python
#As we can see, taking a larger value of N is what makes the histogram look more like a bell curve.

#Let's now try changing "size", the number of samples taken every time
toss(10000,1000,100)
```

```python
toss(10000,1000,50)
```
